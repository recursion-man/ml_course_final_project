```python
import torch
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from PIL import Image
import sys
sys.path.append('..')

train_transform = transforms.Compose([
    transforms.RandomResizedCrop(28, scale=(0.8, 1.0)),  
    transforms.RandomAffine(degrees=10, translate=(0.1, 0.1), shear=5),  
    transforms.ToTensor(),
])
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(28, scale=(0.8, 1.0)), 
    transforms.RandomAffine(degrees=10, translate=(0.1, 0.1), shear=5),  
    transforms.ToTensor(),
])
dataset = datasets.MNIST(root='data', train=True, download=True)

original_image, label = dataset[12]  

original_pil = original_image 

augmented_images = [train_transform(original_pil) for _ in range(6)]

# Plot the original and augmented images
fig, axes = plt.subplots(1, 7, figsize=(14, 4))

# Show original image
axes[0].imshow(original_pil, cmap='gray')  # Use original PIL image
axes[0].set_title("Original")
axes[0].axis('off')

# Show augmented versions
for i in range(6):
    axes[i + 1].imshow(augmented_images[i].squeeze(), cmap='gray')
    axes[i + 1].set_title(f"Aug {i+1}")
    axes[i + 1].axis('off')

plt.suptitle(f'Original and Augmented Variations of MNIST Digit {label}', fontsize=14)
plt.show()
```
